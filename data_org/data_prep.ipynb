```python
import pandas as pd
import os 
```

```python
pth = "../raw_asset_data"
```

```python
if os.path.isdir(pth) == False:
    os.mkdir(pth)
```

```python
data = pd.read_csv ('train.csv')
```

```python
data['Time'] = pd.to_datetime(data['timestamp'], unit='s')
```

```python
asset_details = pd.read_csv('asset_details.csv')
```

```python
data
```

```python
asset_details.head(15)
```

```python
asset_list = []

for each in asset_details['Asset_ID']:
    asset_name = asset_details['Asset_Name'].loc[asset_details['Asset_ID'] == each].values
    asset_name = asset_name[0].replace(' ', '_' )
    
    print(each, asset_name)
    
    asset_df = pd.DataFrame(data.loc[data['Asset_ID'] == each ])
    asset_df.dropna(inplace=True)
    # asset_df.reset_index(drop=True, inplace=True)
    
    asset_list.append(asset_df)
    asset_df.to_csv(f'{pth}/{each}_{asset_name}.csv')
    
```
