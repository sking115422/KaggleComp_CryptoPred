```python
data_returns = pd.DataFrame()


for each in asset_details['Asset_ID']:
    
    df_temp = pd.DataFrame()
    df_temp = data.loc[data['Asset_ID'] == each]
    log_returns_list = []
    
    try:
        for i in range(0, df_temp.shape[0]):
            log_returns_list.append(math.log(df_temp['Close'].iloc[i+16] / df_temp['Close'].iloc[i+1]))
    except:
        log_returns_list.append(None)
        
    if len(log_returns_list) != df_temp.shape[0]:
        diff = abs(len(log_returns_list) - df_temp.shape[0])
        for j in range(0, diff):
            log_returns_list.append(None)
            
    df_temp['LogReturns'] = log_returns_list
    
    data_returns = data_returns.append(df_temp)
    
    
    
```

```python
data_returns.head(100)
```

```python
data_returns = data_returns.sort_values(by ='timestamp', ascending = 1)
```

```python
data_returns.tail(25)
```

```python
wamr_df = pd.DataFrame()

tot_uniq = data_returns['timestamp'].unique()

for each in tot_uniq:
    
    print (each)
    
    df_temp = pd.DataFrame()
    df_temp = data_returns.loc[data_returns['timestamp'] == each]
    
    df_temp = df_temp.sort_values(by ='Asset_ID', ascending = 1)
    
    wamr_list = []
    
    num = 0
    den = 0

    for i in range(0, df_temp.shape[0]):
        
        cur_weight = df_temp['Asset_ID'].iloc[i]
        
        num = num + df_temp['LogReturns'].iloc[i] * asset_details ['Weight'].iloc[cur_weight]
        
        den = den + cur_weight
        
    wamr_value = num/den
    
    for i in range(0, df_temp.shape[0]):
        wamr_list.append(wamr_value)
        
    if len(wamr_list) != df_temp.shape[0]:
        diff = abs(len(wamr_list) - df_temp.shape[0])
        for j in range(0, diff):
            wamr_list.append(None)
            
    df_temp['WAMR'] = wamr_list
    
    wamr_df = wamr_df.append(df_temp)
```
